```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Leer los datos
df = pd.read_csv('users.csv')

# Filtrar personas mayores de 50 a침os
mayores_50 = df[df['edad'] > 50]

# Obtener nombres por g칠nero
masculinos = mayores_50[mayores_50['genero'] == 'male']['nombre']
femeninos = mayores_50[mayores_50['genero'] == 'female']['nombre']

# Calcular totales
total_mayores_50 = len(mayores_50)
masculinos_count = len(masculinos)
femeninos_count = len(femeninos)

# Imprimir resultados
print("Total de personas mayores a 50 a침os:\n", total_mayores_50)
print("Masculinos mayores de 50 a침os:\n", masculinos_count)
print("Femeninos mayores de 50 a침os:\n", femeninos_count)

# Configurar estilo de Seaborn
sns.set(style="whitegrid")

# Crear gr치fica de barras con Seaborn
plt.figure(figsize=(8, 6))
ax = sns.countplot(
    x='genero', 
    data=mayores_50,
    palette={'male': 'blue', 'female': 'pink'}
)

# Personalizar la gr치fica
plt.title('Personas mayores de 50 a침os por g칠nero', fontsize=14)
plt.xlabel('G칠nero')
plt.ylabel('Cantidad')
plt.xticks(ticks=[0, 1], labels=['Masculinos', 'Femeninos'])

# Mostrar valores en las barras
for p in ax.patches:
    ax.annotate(
        f'{int(p.get_height())}', 
        (p.get_x() + p.get_width() / 2., p.get_height()),
        ha='center', va='center', fontsize=12, color='black', xytext=(0, 5),
        textcoords='offset points'
    )

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar datos
df = pd.read_csv("users.csv")

# Filtrar top 20 usuarios m치s viejos del Reino Unido
top_20_uk = df[df["pais"] == "United Kingdom"].sort_values(by="edad", ascending=False).head(20)

# Separar por g칠nero
hombres = top_20_uk[top_20_uk["genero"] == "male"]
mujeres = top_20_uk[top_20_uk["genero"] == "female"]

print("Hombres:\n", hombres[["nombre", "edad"]])
print("\nMujeres:\n", mujeres[["nombre", "edad"]])

# Configurar estilo de Seaborn
sns.set(style="whitegrid")
plt.figure(figsize=(12, 6))

# Crear gr치fica de barras con Seaborn
ax = sns.barplot(
    x='nombre',
    y='edad',
    hue='genero',
    data=top_20_uk,
    palette={'male': 'blue', 'female': 'pink'},
    order=top_20_uk.sort_values('edad', ascending=False)['nombre']  # Mantener orden original
)

# Personalizar gr치fico
plt.title("Top 20 usuarios m치s viejos del Reino Unido por g칠nero", fontsize=14)
plt.xlabel("Nombre")
plt.ylabel("Edad")
plt.xticks(rotation=90)
plt.legend(title='G칠nero')

# Ajustar espacio y mostrar
plt.tight_layout()
plt.show()
```
